As we can see that our dataset is completely unbalanced and if we do modelling, our model might get biased and give the proper accuracy with these imbalanced dataset. So first we will try to balance the dataset by using UnderSampling method.

```python
#splitting our dataset into independent and dependent features

X = data.drop(['Class'], axis = 1)
X.head()
X.shape

Y = data['Class']
Y.head()
Y.shape
```

```python
#pip install -U imbalanced-learn
```

```python
#UnderSampling
from imblearn import under_sampling
from imblearn.under_sampling import NearMiss
```

```python
#Implementing UnderSampling to handle imbalaced dataset
nm = NearMiss()
X_1, Y_1 = nm.fit_sample(X,Y)
X_1.shape
Y_1.shape
```

```python
X_1.head()
Y_1.head()
```

```python
plt.figure(figsize = (8,6))
sns.countplot(Y_1, hue = Y_1);
```

As from the above plot we can say that our dataset is now balanced and we can now apply any classification algorithm. So we are applying Logistic regression.

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X_1, Y_1, test_size = 0.20, random_state = 40)

print("x_train : ", x_train.shape)
print("x_test : ", x_test.shape)
print("y_train : ", y_train.shape)
print("y_test : ", x_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(x_train, y_train )
```

```python
y_predicted = model.predict(x_test)
y_predicted
```

```python
model.score(x_test, y_test)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

cm = confusion_matrix(y_test, y_predicted)
print("confusion_matrix : ")
print(cm)
print("accuracy_score : ", accuracy_score(y_test, y_predicted))
print(classification_report(y_test, y_predicted))
```
