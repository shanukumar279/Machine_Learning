### Let's convert Reviews column from categorical column to numerical column

```python
type(data['Reviews'])
```

```python
data['Reviews'] = pd.to_numeric(data['Reviews'],errors = 'coerce' )
```

```python
data.info()
```

### Let's convert Installs column from categorical column to numerical column

```python
data['Installs'].unique()
```

```python
data['Installs'] = data['Installs'].apply(lambda x : x.replace(',',''))
```

```python
data['Installs'].unique()
```

```python
data['Installs'] = data['Installs'].apply(lambda x : int(x.replace('+','')))
```

```python
data['Installs'].unique()
```

```python
data.info()
```

### Let's convert Reviews column from categorical column to numerical column

```python
data['Price'].unique()
```

```python
data['Price'] = data['Price'].apply(lambda x : x.replace('$',''))
```

```python
data['Price'].unique()
```

```python
data['Price'] = pd.to_numeric(data['Price'], errors = 'coerce')
```

```python
data['Price'].unique()
```

```python
data.info()
```

```python
data.head()
```

```python
data['Category'].unique()
```

```python
x = data.groupby(['Category']).Rating.agg(['mean'])
x
y = data.groupby(['Category']).Installs.agg(['mean'])
y
z = data.groupby(['Category']).Price.agg([np.sum])
z
```

```python
plt.subplot(3,1,1)
plt.plot(x, 'o', linestyle= 'dashed')
plt.xticks(rotation = 90)
plt.title('Category wise Rating', size = 15)
plt.xlabel("Category", size =12)
plt.ylabel("Rating", size =12)
plt.grid()
plt.subplot(3,1,2)
plt.plot(y, 'o', linestyle= 'dashed', color = 'red')
plt.xticks(rotation = 90)
plt.title('Category wise Reviews', size = 15)
plt.xlabel("Category", size =12)
plt.ylabel("Reviews", size =12)
plt.grid()
plt.subplot(3,1,3)
plt.plot(z, 'o', linestyle= 'dashed', color = 'green')
plt.xticks(rotation = 90)
plt.title('Category wise Price', size = 15)
plt.xlabel("Category", size =12)
plt.ylabel("Price", size =12)
plt.grid()
plt.subplots_adjust(right = 2.2, bottom = -2.5, hspace = 1);
```
